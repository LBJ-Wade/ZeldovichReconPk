```python
import sys
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load Example Power Spectrum
klin, plin= np.loadtxt('./ds14_a_1.0000_pk.dat').T
```

```python
# Make the cross spectra between matter, displaced and shifted fields
# Assume R = 15 and Gaussian smoothing

R = 15.0

Sk = np.exp(-(klin*R)**2/2)
Skm1 = - np.expm1(-(klin*R)**2/2)

pmm = plin
pdm = Skm1 * plin
psm = Sk * plin
pdd = Skm1**2 * plin
pds = Sk*Skm1 * plin
pss = Sk**2 * plin

plins = np.array([pmm,pdm,psm,pdd,pds,pss]).T
```

```python
# Import Libraries
from zeldovich_rsd_recon import Zeldovich_Recon_RSD
```

```python
# Make Instance
zelda = Zeldovich_Recon_RSD(klin,plins)
```

```python
# We can plot the various two point functions:
fig = plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.semilogx(zelda.qv, zelda.Xlins['mm'],'k',label=r'$X^{mm}$')
plt.semilogx(zelda.qv, zelda.Xlins['dd'],label=r'$X^{dd}$')
plt.semilogx(zelda.qv, zelda.Xlins['ds'],label=r'$X^{ds}$')
plt.semilogx(zelda.qv, zelda.Xlins['ss'],label=r'$X^{ss}$')
plt.legend()
plt.xlim(1e-2,1e4)
plt.xlabel(r'$q\quad$ [$h^{-1}$ Mpc]')

plt.subplot(1,2,2)
plt.semilogx(zelda.qv, zelda.Ylins['mm'],'k',label=r'$Y^{mm}$')
plt.semilogx(zelda.qv, zelda.Ylins['dd'],label=r'$Y^{dd}$')
plt.semilogx(zelda.qv, -zelda.Ylins['ds'],'--',label=r'$Y^{ds}$')
plt.semilogx(zelda.qv, zelda.Ylins['ss'],label=r'$Y^{ss}$')
plt.legend(loc='upper left')
plt.xlim(1e-2,1e4)
plt.xlabel(r'$q\quad$ [$h^{-1}$ Mpc]')

plt.show()
```

```python
fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.semilogx(zelda.qv,-zelda.qv*zelda.Ulins['mm'],'k')
plt.semilogx(zelda.qv,-zelda.qv*zelda.Ulins['dm'],label=r'$-q\ U^{dm}$')
plt.semilogx(zelda.qv,+zelda.qv*zelda.Ulins['sm'],'--',label=r'$-q\ U^{sm}$')
plt.xlim(0.001,3000)
plt.ylim(-1,35)
plt.legend()
plt.xlim(1e-2,1e4)
plt.xlabel(r'$q\quad$ [$h^{-1}$ Mpc]')

plt.subplot(1,2,2)
plt.semilogx(zelda.qv,-zelda.qv*zelda.Ulins['mm'],'k',label=r'$-q\ U^{mm}$')
plt.semilogx(zelda.qv,-zelda.qv*zelda.Ulins['dd'],label=r'$-q\ U^{dd}$')
plt.semilogx(zelda.qv,zelda.qv*zelda.Ulins['ds'],'--',label=r'$-q\ U^{ds}$')
plt.semilogx(zelda.qv,-zelda.qv*zelda.Ulins['ss'],label=r'$-q\ U^{ss}$')
plt.xlim(0.001,3000)
plt.ylim(-1,35)
plt.legend()
plt.xlim(1e-2,1e4)
plt.xlabel(r'$q\quad$ [$h^{-1}$ Mpc]')

plt.show()

```
